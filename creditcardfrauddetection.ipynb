```python
X = data.drop("Class", axis=1)
y = data["Class"]
```

```python
y.value_counts()
```

## Balance data through SMOTE

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE()

X_resampled, y_resampled = sm.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

## Standardize Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
```

```python
pd.DataFrame(X_scaled).head()
```

```python
y_scaled = y_resampled.astype(bool).values
```

```python
y_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
# Decision Tree
dt_classifier = DecisionTreeClassifier()
# dt_classifier.fit(X_scaled, y_scaled)
# Random Forest
rf_classifier = RandomForestClassifier()
# rf_classifier.fit(X_scaled, y_scaled)
# XGBoost
xgb_classifier = XGBClassifier()
# xgb_classifier.fit(X_scaled, y_scaled)
# Extra Trees
et_classifier = ExtraTreesClassifier()
# et_classifier.fit(X_scaled, y_scaled)
# LightGBM
lgbm_classifier = LGBMClassifier()
# lgbm_classifier.fit(X_scaled, y_scaled)
```

```python
from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=5)

dt_results = cross_val_score(dt_classifier, X_scaled, y_scaled, scoring="f1", cv=kf)
print(f"\nDecision Tree:\n{dt_results}\n")
rf_results = cross_val_score(rf_classifier, X_scaled, y_scaled, scoring="f1", cv=kf)
print(f"\nRandom Forest:\n{rf_results}\n")
xgb_results = cross_val_score(xgb_classifier, X_scaled, y_scaled, scoring="f1", cv=kf)
print(f"\nXGBoost:\n{xgb_results}\n")
et_results = cross_val_score(et_classifier, X_scaled, y_scaled, scoring="f1", cv=kf)
print(f"\nExtra Trees:\n{et_results}\n")
lgbm_results = cross_val_score(lgbm_classifier, X_scaled, y_scaled, scoring="f1", cv=kf)
print(f"\nLightGBM:\n{lgbm_results}\n")
```
